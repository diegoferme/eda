# Access through files: binary formats

The following is a simple example of data access using Python files, using the facilities provided by the **Pandas** library.

The input file corresponds to traffic accidents in the city of Madrid in 2019 and can be downloaded from the following page.

```python
import pandas as pd
```

```python
accidentes = pd.read_csv('./2019_Accidentalidad.csv', 
                        delimiter = ";",
                        encoding = 'utf-8',
                        parse_dates = ["fecha"],
                        na_values = ['-', 'Desconocido'])
```

```python
accidentes[:5]
```

```python
columnas_dropped = ['cod_distrito', 'lesividad', 'coordenada_x_utm', 
                    'coordenada_y_utm', 'positiva_alcohol', 'positiva_droga']
```

```python
accidentes.drop(columnas_dropped, axis = 1, inplace = True)
```

```python
accidentes[:5]
```

```python
accidentes.to_parquet('./accidentes.parquet')
```
