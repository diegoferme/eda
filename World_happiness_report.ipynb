## Context

The World Happiness Report is a landmark survey on the state of global happiness. The first report was published in 2012, the second in 2013, the third in 2015, and the fourth in the 2016 update. The World Happiness 2017, which ranks 155 countries according to their happiness levels, was released at the United Nations at an event celebrating the International Day of Happiness on March 20. The report continues to gain worldwide recognition as governments, organizations, and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts from various fields (economics, psychology, survey analysis, national statistics, health, public policy, and more) describe how well-being measurements can be effectively used to assess the progress of nations.

## Content

The happiness scores and rankings use data from the Gallup World Poll. The scores are based on responses to the main life evaluation question asked in the survey. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their current lives on that scale. The scores come from nationally representative samples for the years 2013-2016 and use Gallup weights to make the estimates representative. The columns that follow the happiness score estimate the extent to which each of the six factors:

- Economic production,

- Social support,

- Life expectancy,

- Freedom,

- Absence of corruption, and

- Generosity

contribute to making life evaluations higher in each country than in dystopia, a hypothetical country that has values equal to the world's lowest national averages for each of the six factors. These columns do not impact the total score reported for each country but do explain why some countries rank higher than others.

## What is Dystopia?
Dystopia is an imaginary country that has the world's least happy people. The purpose of establishing dystopia is to have a benchmark with which all countries can be favorably compared (no country performs worse than dystopia) in terms of each of the six key variables, allowing each bar to be of positive width. The lowest observed values for the six key variables, therefore, characterize dystopia. Life would be very unpleasant in a country with the world's lowest incomes, lowest life expectancy, lowest generosity, most corruption, least freedom, and least social support, in contrast to utopia.

### 4. I check the number of nulls in each column, as well as their percentage.

```python
def show_nulls(df):
    total_nan = df.isnull().sum()
    perc_nan = total_nan/len(df)*100
    return total_nan, perc_nan
```

```python
print("*QUANTITY of null data per column")
print(show_nulls(df=df)[0])
print("----------------------------------")
print("*PERCENTAGE of null data per column")
print(show_nulls(df=df)[1])
```

### 5. I change the null values of the “Lower Confidence Interval” and “Upper Confidence Interval” columns to a random number between the minimum and maximum value of the same column.

First we show the maximum and minimum values of the columns *Lower Confidence Interval* and *Upper Confidence Interval*.

```python
LCI_max = df["Lower Confidence Interval"].max()
LCI_min = df["Lower Confidence Interval"].min()
UCI_max = df["Upper Confidence Interval"].max()
UCI_min = df["Lower Confidence Interval"].min()
print("LOWER CONFIDENCE INTERVAL:")
print("- The maximum value is", LCI_max)
print("- The minimum value is", LCI_min)
print("-"*20)
print("UPPER CONFIDENCE INTERVAL:")
print("- The maximum value is", UCI_max)
print("- The minimum value is", UCI_min)
```

The random values for the lower and upper confidence intervals are as follows

```python
import random
random_LCI = random.uniform(LCI_max,LCI_min)
random_UCI = random.uniform(UCI_max,UCI_min)
print("- The random value for the lower confidence interval is", random_LCI)
print("- The random value for the upper confidence interval is", random_UCI)
```

I replace the null values by these values

```python
df["Lower Confidence Interval"] = df['Lower Confidence Interval'].fillna(round((random_LCI),3))
df["Upper Confidence Interval"] = df['Upper Confidence Interval'].fillna(round((random_UCI),3))
```
